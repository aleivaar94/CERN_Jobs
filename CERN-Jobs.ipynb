```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
jobs = {'Title': ['Develop new medical applications',
                  'Make science more accessible',
                  'Find new ways to describe the Universe',
                  'Implement artificial intelligence',
                  'Manage Big Data',
                  'Look for new particles',
                  'Do another job',
                  'Design novel machines',
                  'Be the Director-General',
                  'Plan for future research',
                  'Work with antimatter',
                  'Make CERN more environmentally friendly'],
        'count': [749, 716, 1046, 1018, 805, 2315, 897, 845, 859, 1129, 901, 988]
       }

df = pd.DataFrame(jobs)
df = df.sort_values('count', ascending=False)
df
```

```python
# Set figure size
dpi = 200
pixel_width, pixel_height = 1080,1080
fig_size_width, fig_size_height = pixel_width / dpi, pixel_height / dpi
plt.figure(figsize=(fig_size_width, fig_size_height), dpi=dpi)

# Change the global font family
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']  # Specify your preferred font

# Create color palette
color = ['#943126'] + (['#004c6d'] * (len(df)-1))  # Change the color of the first bar

# Create the barplot
sns.barplot(x='count', y='Title', data=df, 
            orient='h', palette=color, edgecolor='black',
            linewidth=1.2)
plt.title('At CERN I want to...', loc='left', fontsize=20, fontweight='bold'	)
plt.xlabel('')  # Remove the x-axis title
plt.ylabel('')  # Remove the y-axis title

ax = plt.gca() # Get the current Axes that the plot is on
ax.spines['top'].set_visible(False) # Remove the top border
ax.spines['right'].set_visible(False) # Remove the right border

ax.grid(True, which='both', axis='x', color='gray',
        linestyle='--', linewidth=0.5, alpha=0.5) # Add the grid lines
ax.tick_params(axis='y', labelsize=10)
ax.tick_params(axis='x', labelsize=12)


plt.savefig('cern_jobs.png', dpi=dpi, bbox_inches='tight')

plt.show()
```
